```python
!pip install yfinance plotly --quiet

import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries loaded")
```

```python
# SAP vs. Major Competitors in ERP/Cloud
tickers = {
    "SAP": "SAP SE",
    "ORCL": "Oracle",
    "CRM": "Salesforce",
    "MSFT": "Microsoft"
}

# Last 5 years — Pre- and post-SAP cloud transformation snapshot
datos = yf.download(list(tickers.keys()), period="5y", auto_adjust=True)['Close']
retornos = (datos / datos.iloc[0] - 1) * 100

print("✅ Stock data downloaded")
print(f"Period: {datos.index[0].date()} → {datos.index[-1].date()}")
print(f"\nCurrent prices:")
for ticker, nombre in tickers.items():
    print(f"  {nombre}: ${datos[ticker].iloc[-1]:.2f}")
```

```python
# Key Financial Metrics from SAP
sap = yf.Ticker("SAP")
info = sap.info

print("📊 SAP SE — KEY FINANCIAL METRICS")
print("="*50)

metricas = {
    "Market Cap": info.get('marketCap', 'N/A'),
    "Revenue (TTM)": info.get('totalRevenue', 'N/A'),
    "Gross Margin": info.get('grossMargins', 'N/A'),
    "Operating Margin": info.get('operatingMargins', 'N/A'),
    "P/E Ratio": info.get('trailingPE', 'N/A'),
    "52W High": info.get('fiftyTwoWeekHigh', 'N/A'),
    "52W Low": info.get('fiftyTwoWeekLow', 'N/A'),
}

for k, v in metricas.items():
    if isinstance(v, float):
        if v > 1000:
            print(f"  {k}: ${v/1e9:.1f}B")
        elif v < 1:
            print(f"  {k}: {v*100:.1f}%")
        else:
            print(f"  {k}: {v:.2f}")
    else:
        print(f"  {k}: {v}")
```

```python
# SAP's historical annual revenue (public data from annual reports)
# Source: SAP Annual Reports 2019-2023
sap_revenue = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Cloud Revenue (€B)': [6.93, 8.08, 9.42, 11.43, 13.66],
    'Software Licenses (€B)': [3.84, 2.96, 2.70, 2.29, 1.35],
    'Services (€B)': [4.27, 3.84, 3.78, 4.22, 4.42],
    'Total Revenue (€B)': [27.55, 27.34, 27.84, 30.87, 31.21],
    'Cloud % of Total': [25.2, 29.6, 33.8, 37.0, 43.8]
})

print("📊 SAP REVENUE BREAKDOWN (2019-2023)")
print(sap_revenue.to_string(index=False))
print(f"\n🔑 Cloud revenue grew {((13.66/6.93)-1)*100:.1f}% from 2019 to 2023")
print(f"🔑 Cloud as % of total: {sap_revenue['Cloud % of Total'].iloc[0]}% → {sap_revenue['Cloud % of Total'].iloc[-1]}%")
```

```python
# SAP's Presence in Latin America
# Source: SAP Annual Reports, IDC, SAP.com/latam

latam_data = pd.DataFrame({
    'Country': ['Brazil', 'Mexico', 'Argentina', 'Colombia', 'Chile', 'Peru', 'Others'],
    'SAP Customers (est.)': [8500, 4200, 2100, 1800, 1500, 900, 2000],
    'Market Share ERP (%)': [38, 32, 35, 28, 33, 25, 22],
    'Cloud Adoption (%)': [45, 38, 31, 35, 42, 28, 25],
    'YoY Growth (%)': [12, 9, 8, 11, 13, 10, 7]
})

print("📊 SAP LATAM PRESENCE (2023 estimates)")
print(latam_data.to_string(index=False))
print(f"\n🇦🇷 Argentina: ~{latam_data[latam_data['Country']=='Argentina']['SAP Customers (est.)'].values[0]:,} customers")
print(f"🇧🇷 Brazil: ~{latam_data[latam_data['Country']=='Brazil']['SAP Customers (est.)'].values[0]:,} customers — largest LATAM market")
```

```python
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "SAP vs Competitors — 5Y Stock Return (%)",
        "SAP Cloud Revenue Growth (€B)",
        "SAP LATAM Market Share by Country (%)",
        "SAP LATAM Cloud Adoption (%)"
    ),
    vertical_spacing=0.15,
    horizontal_spacing=0.1
)

# Colors by company
colores_comp = {
    "SAP": "#1F77B4",
    "ORCL": "#FF7F0E",
    "CRM": "#2CA02C",
    "MSFT": "#9467BD"
}

# Panel 1: 5-Year Stock Market Return (Weekly, for simplicity)
datos_weekly = datos.resample('W').last()
retornos_weekly = (datos_weekly / datos_weekly.iloc[0] - 1) * 100

for ticker, nombre in tickers.items():
    fig.add_trace(go.Scatter(
        x=retornos_weekly.index,
        y=retornos_weekly[ticker].round(2),
        name=nombre,
        line=dict(color=colores_comp[ticker], width=2),
        showlegend=True
    ), row=1, col=1)

# Panel 2: Cloud revenue growth
fig.add_trace(go.Bar(
    x=sap_revenue['Year'],
    y=sap_revenue['Cloud Revenue (€B)'],
    name='Cloud Revenue',
    marker_color='#1F77B4',
    text=sap_revenue['Cloud Revenue (€B)'].round(1),
    textposition='outside',
    showlegend=False
), row=1, col=2)

fig.add_trace(go.Scatter(
    x=sap_revenue['Year'],
    y=sap_revenue['Cloud % of Total'],
    name='Cloud % of Total',
    line=dict(color='orange', width=3, dash='dot'),
    yaxis='y4',
    showlegend=False
), row=1, col=2)

# Panel 3: Market share LATAM
fig.add_trace(go.Bar(
    x=latam_data['Country'],
    y=latam_data['Market Share ERP (%)'],
    marker_color='#1F77B4',
    text=latam_data['Market Share ERP (%)'].astype(str) + '%',
    textposition='outside',
    showlegend=False
), row=2, col=1)

# Panel 4: Cloud adoption LATAM
fig.add_trace(go.Bar(
    x=latam_data['Country'],
    y=latam_data['Cloud Adoption (%)'],
    marker_color='#2CA02C',
    text=latam_data['Cloud Adoption (%)'].astype(str) + '%',
    textposition='outside',
    showlegend=False
), row=2, col=2)

fig.update_layout(
    title=dict(
        text="SAP Business Analysis — Global Performance & LATAM Expansion",
        font=dict(size=18), x=0.5
    ),
    height=750,
    template="plotly_white",
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)

fig.write_html("sap_dashboard.html", include_plotlyjs='cdn')
fig.show()
print("✅ SAP dashboard saved")
```

```python
# SAP vs. Competitors Comparison Chart
competidores_kpi = []
for ticker, nombre in tickers.items():
    t = yf.Ticker(ticker)
    info = t.info
    competidores_kpi.append({
        'Company': nombre,
        'Market Cap ($B)': round(info.get('marketCap', 0) / 1e9, 1),
        'Revenue ($B)': round(info.get('totalRevenue', 0) / 1e9, 1),
        'Gross Margin (%)': round(info.get('grossMargins', 0) * 100, 1),
        '5Y Return (%)': round(retornos[ticker].iloc[-1], 1)
    })

kpi_df = pd.DataFrame(competidores_kpi).sort_values('Market Cap ($B)', ascending=False)

fig2 = go.Figure(data=[go.Table(
    header=dict(
        values=[f'<b>{c}</b>' for c in kpi_df.columns],
        fill_color='#1F4E79',
        font=dict(color='white', size=13),
        align='center', height=35
    ),
    cells=dict(
        values=[kpi_df[c] for c in kpi_df.columns],
        fill_color=[['#EBF1F9' if i%2==0 else 'white'
                     for i in range(len(kpi_df))]],
        font=dict(size=12), align='center', height=30
    )
)])

fig2.update_layout(
    title=dict(text="SAP vs Competitors — Key Financial Metrics",
               font=dict(size=16), x=0.5),
    height=280
)

fig2.write_html("sap_competitors_kpi.html", include_plotlyjs='cdn')
fig2.show()
print("✅ KPI table saved")
```
