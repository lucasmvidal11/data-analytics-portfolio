```python
import pandas as pd
import numpy as np
import requests
import warnings
warnings.filterwarnings("ignore")

print("OK")
```

```python
df = pd.read_csv("../data/credit_risk_dataset.csv")

print(f"Shape: {df.shape}")
print(f"\nColumnas: {list(df.columns)}")
print(f"\nPrimeras filas:")
print(df.head())
```

```python
print("Distribución de defaults:")
print(df["loan_status"].value_counts())
print(f"\nTasa de default: {df['loan_status'].mean()*100:.1f}%")

print("\nNulos por columna:")
print(df.isnull().sum())

print("\nEstadísticas básicas:")
print(df[["person_income", "loan_amnt", "loan_int_rate", "loan_percent_income"]].describe().round(2))
```

```python
# Inflación anual Argentina (INDEC - datos oficiales verificados)
inflation_annual = pd.DataFrame({
    "year":             [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "inflation_annual": [26.9, 41.0, 24.8, 47.6, 53.8, 36.1, 50.9, 94.8, 211.4, 117.8]
})

print("Inflación anual Argentina (INDEC):")
print(inflation_annual.to_string(index=False))
```

```python
# Tasa de política monetaria Argentina (BCRA - datos históricos verificados)
rates_df = pd.DataFrame({
    "year":         [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "lending_rate": [27.0, 30.5, 28.75, 60.0, 63.0, 38.0, 38.0, 75.0, 133.0, 110.0]
})

print("Tasa de política monetaria Argentina (BCRA):")
print(rates_df.to_string(index=False))
```

```python
macro_df = inflation_annual.merge(rates_df, on="year", how="inner")
macro_df = macro_df[macro_df["year"] >= 2015]

print("Macro data Argentina:")
print(macro_df.to_string(index=False))

macro_df.to_csv("../data/argentina_macro.csv", index=False)
df.to_csv("../data/credit_risk_clean.csv", index=False)
print("\nGuardado OK")
```
